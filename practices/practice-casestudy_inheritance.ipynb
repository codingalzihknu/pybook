그리고 `inheritance.py` 는 `Sprite` 클래스를 상속하는 `Character` 클래스를 추가로 선언한다.

```python
# Sprite 클래스 상속 
class Character(Sprite):
    def __init__(self, x, y, width, height, image, jump=False):
        super().__init__(x, y, width, height, image)
        self.jump = jump 

    def hop(self, distance=300):
        if self.jump == True:
            self.x += distance
```
